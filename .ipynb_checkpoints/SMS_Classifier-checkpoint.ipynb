### 1. Introduccíon

* **SMS y SPAM. Que son?**

Incluso en estos tiempos en que la tecnología avanza a pasos agigantados, el SMS sigue siendo una plataforma de comunicación muy utilizada. Sólo se necesita el número de teléfono del destinatario, y podemos enviar un mensaje que llega casi instantáneamente, sin utilizar datos.

*SPAM* se define como mensajes no solicitados, generalmente comerciales (como correos electrónicos, mensajes de texto o publicaciones en Internet), enviados a un gran número de destinatarios o publicados en un gran número de lugares.

El **SMS spamming** es una antigua práctica de marketing y sigue siendo un problema que causa molestias a los consumidores de todo el mundo. Los mensajes SPAM procedentes de SMS y aplicaciones de mensajería están cada vez más extendidos. Según los últimos datos del sector, más de la mitad de los usuarios de mensajes de texto de todo el mundo reciben ya al menos un mensaje de SPAM semanal. Y lo que es más alarmante, más de una cuarta parte recibe spam todos los días con mensajes no solicitados.

El spam puede llenar tu bandeja de entrada de mensajes no solicitados y a veces no son sólo anuncios, a veces son incluso una *amenaza para la seguridad*.

* **Descripción del problema y estructura de los datos.**

Mantener la bandeja de entrada sin mensajes no deseados, evitando anuncios no solicitados o mensajes que incluyan amenazas que puedan infectar tu teléfono o robar tus datos privados, es un problema que a día de hoy seguimos teniendo en nuestros dispositivos móviles.

Una aplicación capaz de filtrar los SMS entrantes, descartando aquellos que son SPAM, evitaría más de un problema a los usuarios. Utilizando algoritmos y técnicas de aprendizaje automático como [**_TLN_**](https://es.wikipedia.org/wiki/Tratamiento_de_lenguajes_naturales), se puede desarrollar un modelo de machine learning para este fin. 

Utilizando un [dataset](http://www.dt.fee.unicamp.br/~tiago/smsspamcollection/) con más de 5000 SMS recopilados para la investigación del SPAM en telefonía móvil, analizaremos los datos y desarrollaremos un modelo de aprendizaje automático para clasificar los SMS. El conjunto de datos es un archivo CSV, y tiene 5 columnas; la primera los valores **_spam/ham_**, que será el valor que queremos clasificar. Y las 4 restantes con el texto del SMS.

Como ya hemos explicado, etiquetar como SPAM un SMS que no lo es,  no debe suceder, pero etiquetar demasiados SMS spam como mensajes legítimos tampoco es el objetivo. Por eso mediremos la efectividad del modelo con las métricas **_Precisión, Recall y F1_**.

Luego, dentro de este mismo _notebook_, hay creadas 2 clases más, _basicFeat_ y _txtNormalizer_. Estas clases son las encargadas de normalizar el cuerpo de los SMS, y de crear nuevas propiedades a cada uno de ellos, para ayudar a la tarea de clasificarlos como _SPAM_ o _HAM_.

[Aquí](https://github.com/JV-Peluso/SMS_Classifier/blob/main/classHelp.txt) está la descripción de cada librería y clase (en inglés), por si os interesa verlos a fondo.

```python
class basicFeat(BaseEstimator, TransformerMixin):
    
    ''' Clase que crea nuevas propiedades basada en el cuerpo del SMS '''
    
    def __init__(self):
        self.col1 = 'smsLen'
        self.col2 = 'smsWords'
        self.col3 = 'smsUpper'
        self.col4 = 'smsSpecChar'
        self.col5 = 'smsWordLen'
        self.colsNames = ['smsLen', 'smsWords','smsUpper', 'smsSpecChar','smsWordLen']
        self.textF = 'SMS'
        self.targF = 'SPAM'
        
    def fit(self, X, y=None):
        return self

    def transform(self, X):
        Z = None
        if isinstance(X, pd.Series) : 
            Z = pd.DataFrame(X, columns = [self.textF])
        else:
            Z = X.copy()
        return self.createFeat(Z)

    def get_feature_names(self):
        return self.colsNames
    
    def createFeat(self, X):       
        
        DF = pd.DataFrame()
        self.lenCount(X, DF, self.textF, self.col1)
        self.wordCount(X, DF, self.textF, self.col2)
        self.wordUpper(X, DF, self.textF, self.col3)
        self.specCharCount(X, DF, self.textF, self.col4)
        self.wordMeanLen(X, DF, self.textF, self.col5)
        return DF.to_numpy()
        
    def createCol(self, df):
        
        self.lenCount(df, df, self.textF, self.col1)
        self.wordCount(df, df, self.textF, self.col2)
        self.wordUpper(df, df, self.textF, self.col3)
        self.specCharCount(df, df, self.textF, self.col4)
        self.wordMeanLen(df, df, self.textF, self.col5)
            
    def lenCount(self, dfO, dfT, col, colN):
        dfT[colN] = dfO[col].str.len()
    
    def splitWordsC(self, sms):
        words = sms.split()
        return len(words)
        
    def wordCount(self, dfO, dfT, col, colN):
        dfT[colN] = dfO[col].apply(self.splitWordsC)
    
    def splitWords(self, sms):
        words = sms.split()
        return len([word for word in words if word.isupper()])
        
    def wordUpper(self, dfO, dfT, col, colN):
        dfT[colN] = dfO[col].apply(self.splitWords)
        
    def applyRegEx(self, sms):
        regPattern = r"[\sa-zA-Z0-9.,'']"
        return len(sms) - len(re.findall(regPattern, sms))
    
    def specCharCount(self, dfO, dfT, col, colN):
         dfT[colN] = dfO[col].apply(self.applyRegEx)
            
    def wordMeanLen(self, dfO, dfT, col, colN):
        dfT[colN] = dfO[col].apply(self.avgWordLen)
    
    def avgWordLen(self, sms):
        words = sms.split()
        return np.mean([len(word) for word in words])
```

```python
class txtNormalizer(BaseEstimator, TransformerMixin):
    
    ''' Clase que normaliza el texto de los SMS '''
    
    def __init__(self):
        self.nlp = spacy.load('en_core_web_sm')
        self.SW = spacy.lang.en.stop_words.STOP_WORDS
        
    def txtClean(self, sms):
        
        SMS = self.nlp(self.toLower(sms))
        SMS = self.removePunt(SMS)
        SMS = self.removeSW(SMS)
        SMS = self.smsLemma(SMS)
        return ' '.join(SMS)
    
    def toLower(self, sms):
        return sms.lower()
    
    def removePunt(self, sms):
        return [w for w in sms if w.text not in string.punctuation and 
                w.text.isalpha()]
        
    def removeSW(self, sms):
        return [w for w in sms if w.text not in self.SW]
        
    def smsLemma(self, sms):
        return[w.lemma_ for w in sms] 
```

### 5. Desarrollo del modelo

* **Partición de datos y punto de referencia.**

Luego, sólo con los atributos básicos creados previamente para la fase del análisis descriptivo, creamos un modelo básico de *Regresión Logística* y comprobamos su rendimiento con un proceso de validación cruzada, para saber qué resuñtado obtenemos de base, antes de desarrollar modelos más afinados. 

```python
# Dividir los datos usados en el análisis explorativo en entrenamiento/prueba y normalizarlos 
X_train, X_test, y_train, y_test = dh.dataPrep(edaDF, lstFeat, 'SPAM', 0.3, 'SPAM', norm = True)
```

La _**Accuracy**_ (Exactitud) siempre será alta en datos desbalanceados, pero los resultados de **_Precision_**, **_Recall_** y, en consecuencia, de **_F1_** reflejan que el modelo no maneja correctamente la clase minoritaria (_spam_). Por lo tanto, necesitamos añadir mas atributos al modelo.

* **Creación de atributos TLN, selección de modelo y tuneado de hiperparámetros.**

En primer lugar, dividimos el conjunto de datos original (sólo el corpus del SMS y la etiqueta de Spam/No-Spam) en las mismas proporciones que antes, y creamos un _**pipeline**_ con los siguientes pasos:

1. Creación de la matriz de atributos (básicas y [TLN](https://es.wikipedia.org/wiki/Tratamiento_de_lenguajes_naturales)) con FeatureUnion, los pasos internos son:

      * Creación de las atributos básicos.
      * Creación de la matriz _CountVectorizer_, pre-procesando el texto (minúsculas, stopwords, etc.) y limitando el número de propiedades a 2500.
      * Creación de la matriz _TF-IDF_, preprocesando el texto (minúsculas, stopwords, etc.) y limitando el número de propiedades a 2500.
      
      
2. Normalizar los datos con el método _StandardScaler_.
3. Clasificar los SMS con los modelos de machine learning.  

Para cada modelo, ejecutamos el una validación cruzada por 3 distintos valores de los hiperparámetros mas relevantes de cada algoritmo, para encontrar los mejores valores y evitar en un futuro *overfitting* (sobreajuste).

Los algoritmos de aprendizaje automático seleccionados son _**Clasificador Naïve Bayes**_, _**Regresión Logística**_ y _**SVM (Máquinas de Vectores de Soporte)**_. Principalmente, porque los algoritmos lineales son muy sencillos de entrenar y los resultados son interpretables, ya que se pueden extraer fácilmente los coeficientes más importantes del modelo.

```python
# Dividir los datos originales en set de entrenamiento/prueba 
X_train, X_test, y_train, y_test = dh.dataPrep(origDF, 'SMS', 'SPAM', 0.3, 'SPAM')
```

```python
# Crer el FeatureUnion uniendo los atributos básicos con las matrices CV/TF-IDF
basicF = basicFeat()
countV = CountVectorizer(preprocessor = txtN.txtClean, ngram_range = (1,2),
                         max_features = 2500)
tfidf = TfidfVectorizer(preprocessor = txtN.txtClean, ngram_range = (1,2),
                        max_features = 2500)
combFeat = FeatureUnion([('basicF', basicF),
                         ('countV', countV),
                         ('tfidf', tfidf)])
X_features = combFeat.fit(X_train, y_train).transform(X_train)
print("Total atributos (Basicos + CV + TF-IDF) --> {}".format(X_features.shape[1]))
```

Ahora, uno a uno creamos los modelos y les alimentamos con este conjunto de datos, buscando el mejor valor para los hiperparámetros.

* *Clasificador Naïve Bayes*

```python
lstScore = ['accuracy', 'precision', 'recall', 'f1']
stdScaler = StandardScaler(with_mean=False)
alphaVal = [0.02, 0.03, 0.04]
lstResults = []
for item in alphaVal:
    pipeline = Pipeline([('feat', combFeat), ('scaler', stdScaler),
                         ('clf', MultinomialNB(alpha = item))])
    scores = cross_validate(pipeline, X_train, y_train, scoring = lstScore, cv = 3)
    tmpLst = [item]
    for score in lstScore:
        tmpLst.append(round(scores['test_' + score].mean(),4))
    lstResults.append(tmpLst)
    
cols = ['Alpha']
for score in lstScore:
    cols.append(score.capitalize())
df = pd.DataFrame(lstResults, columns=cols)
df.head()
```

* *Regresión Logística*

```python
CVal = [0.1, 0.3, 0.5]
lstResults = []
for item in CVal:
    pipeline = Pipeline([('feat', combFeat), ('scaler', stdScaler),
                         ('clf', LogisticRegression(C = item))])
    scores = cross_validate(pipeline, X_train, y_train, scoring = lstScore, cv = 3)
    tmpLst = [item]
    for score in lstScore:
        tmpLst.append(round(scores['test_' + score].mean(),4))
    lstResults.append(tmpLst)
    
cols = ['C']
for score in lstScore:
    cols.append(score.capitalize())
df = pd.DataFrame(lstResults, columns=cols)
df.head()
```

* *SVM*

```python
CVal = [0.05, 0.1, 0.15]
lstResults = []
for item in CVal:
    pipeline = Pipeline([('feat', combFeat), ('scaler', stdScaler),
                         ('clf', SVC(kernel = 'linear', gamma = 0.01, C = item))])
    scores = cross_validate(pipeline, X_train, y_train, scoring = lstScore, cv = 3)
    tmpLst = ['linear', 0.01, item]
    for score in lstScore:
        tmpLst.append(round(scores['test_' + score].mean(),4))
    lstResults.append(tmpLst)
    
cols = ['Kernel', 'Gamma', 'C']
for score in lstScore:
    cols.append(score.capitalize())
df = pd.DataFrame(lstResults, columns=cols)
df.head()
```

Aqui mostramos la mejor selección de las metricas más altas por cada uno de los modelos tuneados:

Modelo | Accuracy | Precision | Recall | F1 score | Hiperparámetros 
--- | --- | --- | --- | --- | --- 
Naïve Bayes | 0.9527 | 0.7742 | 0.8906 | 0.8266 | alpha: 0.003
**_Regresión Logística_** | **_0.9735_** | **_0.9796_** | **_0.8074_** | **_0.8844_** | **_C : 0.1_**
SVM | 0.9677 | 0.9369 | 0.7987 | 0.8616 | C : 0.05 - gamma : 0.01

El modelo de **_Regresión Logística_** obtuvo las mejores métricas, con una *Accuracy* y *Precision* superiores a 0,97, lo que significa que casi todos los SMS no spam se etiquetaron correctamente, y una puntuación F1 de 0,88, también la más alta entre los demás modelos.

* **Validación de los modelos**

Con los mejores hiperparámetros encontrados para cada algoritmo, ajustamos los modelos con los datos de entrenamiento, y realizamos las predicciones con los datos de prueba para validar su efectividad. En las matrices de confusión podemos ver los resultados:

```python
# Creamos pipelines con los mejores hiperparametros, entrenamos, y 
# validamos los resultados con el set de prueba
stdScaler = StandardScaler(with_mean=False)
pipeNB = Pipeline([('feat', combFeat), ('scaler', stdScaler),
                   ('clf', MultinomialNB(alpha = 0.03))])
pipeLR = Pipeline([('feat', combFeat), ('scaler', stdScaler),
                   ('clf', LogisticRegression(C = 0.1))])
pipeSVM = Pipeline([('feat', combFeat), ('scaler', stdScaler),
                    ('clf', SVC(kernel = 'linear', C = 0.1, gamma = 0.01))])
lstPipe = [pipeNB, pipeLR, pipeSVM]
lstPipeN = ['Naïve Bayes', 'Regresión Logística','SVM']
lstLbl = ['HAM','SPAM']
preds = mh.modelValidation(X_train, y_train, X_test, y_test, lstPipe)

pg.setStyle(style = 'white')
for i in range(len(preds)):
    pg.plotConfMatrix(lstPipeN[i], y_test, preds[i], lstLbl, cmap='GnBu', 
                      figSize = [3.5,2.5], normalize = True, txtFS = 10, legFS = 10)
```

Los resultados coinciden con los obtenidos en el proceso de tuneo y validación. **_Regresión Logística_** vuelve a ser el mejor modelo, obteniendo *Accuracy* incluso mayor y, aunque la puntuación de *Precision* disminuye, las puntuaciones de *Recall* y *F1* aumentan, por lo que interpretamos que el modelo generaliza mejor.

Vemos que la clasificación correcta de los SMS no spam es de un ¡**_99,48_**%! que es lo que teniamos como objetivo principal. Además, sólo el 16.54% de los SMS spam se clasificaron erróneamente como No-Spam, mejorable pero dentro de los límites aceptables.
