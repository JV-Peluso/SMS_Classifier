### 1. Introduccíon

* **SMS y SPAM. Que son?**

Incluso en estos tiempos en que la tecnología avanza a pasos agigantados, el SMS sigue siendo una plataforma de comunicación muy utilizada. Sólo se necesita el número de teléfono del destinatario, y podemos enviar un mensaje que llega casi instantáneamente, sin utilizar datos.

*SPAM* se define como mensajes no solicitados, generalmente comerciales (como correos electrónicos, mensajes de texto o publicaciones en Internet), enviados a un gran número de destinatarios o publicados en un gran número de lugares.

El **SMS spamming** es una antigua práctica de marketing y sigue siendo un problema que causa molestias a los consumidores de todo el mundo. Los mensajes SPAM procedentes de SMS y aplicaciones de mensajería están cada vez más extendidos. Según los últimos datos del sector, más de la mitad de los usuarios de mensajes de texto de todo el mundo reciben ya al menos un mensaje de SPAM semanal. Y lo que es más alarmante, más de una cuarta parte recibe spam todos los días con mensajes no solicitados.

El spam puede llenar tu bandeja de entrada de mensajes no solicitados y a veces no son sólo anuncios, a veces son incluso una *amenaza para la seguridad*.

* **Descripción del problema y estructura de los datos.**

Mantener la bandeja de entrada sin mensajes no deseados, evitando anuncios no solicitados o mensajes que incluyan amenazas que puedan infectar tu teléfono o robar tus datos privados, es un problema que a día de hoy seguimos teniendo en nuestros dispositivos móviles.

Una aplicación capaz de filtrar los SMS entrantes, descartando aquellos que son SPAM, evitaría más de un problema a los usuarios. Utilizando algoritmos y técnicas de aprendizaje automático como [**_NLP_**](https://es.wikipedia.org/wiki/Tratamiento_de_lenguajes_naturales), se puede desarrollar un modelo de machine learning para este fin. 

Utilizando un [dataset](http://www.dt.fee.unicamp.br/~tiago/smsspamcollection/) con más de 5000 SMS recopilados para la investigación del SPAM en telefonía móvil, analizaremos los datos y desarrollaremos un modelo de aprendizaje automático para clasificar los SMS. El conjunto de datos es un archivo CSV, y tiene 5 columnas; la primera los valores **_spam/ham_**, que será el valor que queremos clasificar. Y las 4 restantes con el texto del SMS.

Como ya hemos explicado, etiquetar como SPAM un SMS que no lo es,  no debe suceder, pero etiquetar demasiados SMS spam como mensajes legítimos tampoco es el objetivo. Por eso mediremos la efectividad del modelo con las métricas **_Precisión, Recall y F1_**.

```python
class basicFeat(BaseEstimator, TransformerMixin):
    
    ''' Clase que crea nuevas propiedades basada en el cuerpo del SMS '''
    
    def __init__(self):
        self.col1 = 'smsLen'
        self.col2 = 'smsWords'
        self.col3 = 'smsUpper'
        self.col4 = 'smsSpecChar'
        self.col5 = 'smsWordLen'
        self.colsNames = ['smsLen', 'smsWords','smsUpper', 'smsSpecChar','smsWordLen']
        self.textF = 'SMS'
        self.targF = 'SPAM'
        
    def fit(self, X, y=None):
        return self

    def transform(self, X):
        Z = None
        if isinstance(X, pd.Series) : 
            Z = pd.DataFrame(X, columns = [data.textF])
        else:
            Z = X.copy()
        return self.createFeat(Z)

    def get_feature_names(self):
        return self.colsNames
    
    def createFeat(self, X):       
        
        DF = pd.DataFrame()
        self.lenCount(X, DF, self.textF, self.col1)
        self.wordCount(X, DF, self.textF, self.col2)
        self.wordUpper(X, DF, self.textF, self.col3)
        self.specCharCount(X, DF, self.textF, self.col4)
        self.wordMeanLen(X, DF, self.textF, self.col5)
        return DF.to_numpy()
        
    def createCol(self, df):
        
        self.lenCount(df, df, self.textF, self.col1)
        self.wordCount(df, df, self.textF, self.col2)
        self.wordUpper(df, df, self.textF, self.col3)
        self.specCharCount(df, df, self.textF, self.col4)
        self.wordMeanLen(df, df, self.textF, self.col5)
            
    def lenCount(self, dfO, dfT, col, colN):
        dfT[colN] = dfO[col].str.len()
    
    def splitWordsC(self, sms):
        words = sms.split()
        return len(words)
        
    def wordCount(self, dfO, dfT, col, colN):
        dfT[colN] = dfO[col].apply(self.splitWordsC)
    
    def splitWords(self, sms):
        words = sms.split()
        return len([word for word in words if word.isupper()])
        
    def wordUpper(self, dfO, dfT, col, colN):
        dfT[colN] = dfO[col].apply(self.splitWords)
        
    def applyRegEx(self, sms):
        regPattern = r"[\sa-zA-Z0-9.,'']"
        return len(sms) - len(re.findall(regPattern, sms))
    
    def specCharCount(self, dfO, dfT, col, colN):
         dfT[colN] = dfO[col].apply(self.applyRegEx)
            
    def wordMeanLen(self, dfO, dfT, col, colN):
        dfT[colN] = dfO[col].apply(self.avgWordLen)
    
    def avgWordLen(self, sms):
        words = sms.split()
        return np.mean([len(word) for word in words])
```

```python
class txtNormalizer(BaseEstimator, TransformerMixin):
    
    ''' Clase que normaliza el texto de los SMS '''
    
    def __init__(self):
        self.nlp = spacy.load('en_core_web_sm')
        self.SW = spacy.lang.en.stop_words.STOP_WORDS
        
    def txtClean(self, sms):
        
        SMS = self.nlp(self.toLower(sms))
        SMS = self.removePunt(SMS)
        SMS = self.removeSW(SMS)
        SMS = self.smsLemma(SMS)
        return ' '.join(SMS)
    
    def toLower(self, sms):
        return sms.lower()
    
    def removePunt(self, sms):
        return [w for w in sms if w.text not in string.punctuation and 
                w.text.isalpha()]
        
    def removeSW(self, sms):
        return [w for w in sms if w.text not in self.SW]
        
    def smsLemma(self, sms):
        return[w.lemma_ for w in sms] 
```
